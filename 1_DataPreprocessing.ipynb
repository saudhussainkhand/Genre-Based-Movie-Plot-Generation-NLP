Making New df that contains Genre's (Superhero, Action, Comedy, Drama, Horror, Thriller) and Saving it. This DF will be used while fine tuning all the Text Generation Models used

```python
newdf=pd.DataFrame(columns=["Genre","Plot"], index=range(len(dataset)))
```

```python
for i in range(len(dataset['Genre'])):

    if dataset['Genre'][i]=='superhero':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<superhero>"+dataset['Plot'][i]
    if dataset['Genre'][i]=='action':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<action>"+dataset['Plot'][i]

    if dataset['Genre'][i]=='drama':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<drama>"+dataset['Plot'][i]
    if dataset['Genre'][i]=='thriller':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<thriller>"+dataset['Plot'][i]
    if dataset['Genre'][i]=='horror':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<horror>"+dataset['Plot'][i]
    if dataset['Genre'][i]=='sci_fi':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<sci_fi>"+dataset['Plot'][i]

    if dataset['Genre'][i]=='comedy':

      newdf.loc[i].Genre=dataset['Genre'][i]
      newdf.loc[i].Plot="<comedy>"+dataset['Plot'][i]
```

```python
newdf=newdf.dropna()
```

```python
newdf=newdf.reset_index()
```

```python
newdf.to_csv("cleaneddf.csv")
```
